```python
##Processed financial data for the model to train on

import json
from pathlib import Path

import pandas as pd
from tqdm import tqdm 

Raw_data = Path("data/nasdaq_exteral_data.csv")
Processed_data = Path("data/processed/processed_financial_text.jsonl")

def get_string(row, col, default = ""):
    if col in row and pd.notna(row[col]) :
        value = row[col]
    else:
        value = default
    return str(value)

def build_processed(input_path, output_path, limit, tickers):
    output_path.parent.mkdir(parents= True, exist_ok=True)
    df = pd.read_csv(input_path, nrows=limit)
    if tickers is not None:
        df = df[df["Stock_symbol"].isin(tickers)]
    if limit is not None:
        df = df.head(limit)
    count = 0
    with output_path.open("w", encoding = "utf-8") as f:
        for _, row in tqdm(df.iterrows(), total = len(df)):
            date = get_string(row, "Date")
            title = get_string(row, "Article_title")
            ticker = get_string(row, "Stock_symbol")
            article = get_string(row, "Article")
            print(count)
            print(f"{ticker}")
            count+=1 
            text = (
                f"Ticker: {ticker}\n"
                f"Date: {date}\n"
                f"title: {title}\n"
                f"article = {article}\n"
            ).strip()

            f.write(json.dumps({"text": text}, ensure_ascii=False) + "\n")
tickers = {"AAPL", "GOOGL", "NVDA"}
build_processed(Raw_data, Processed_data, 200000, tickers)

            
             
        
        
```
